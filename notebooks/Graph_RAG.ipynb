```python
import os

BATCH_SIZE = 10  # adjust depending on GPU/memory
BATCH_SAVE_DIR = "./graph_batches"       # stores node batches
INDEX_PERSIST_DIR = "/workspace/graph_store"  # full index storage

# Make sure directories exist
os.makedirs(BATCH_SAVE_DIR, exist_ok=True)
os.makedirs(INDEX_PERSIST_DIR, exist_ok=True)
```

```python
from llama_index.core import PropertyGraphIndex
from llama_index.core.graph_stores import SimplePropertyGraphStore

graph_store = SimplePropertyGraphStore()

index = PropertyGraphIndex(
    nodes=[],               # start empty, insert batch-wise
    property_graph_store=graph_store,
    llm=llm,                # your Ollama LLM instance
    embed_model=embed_model, 
    use_async=False,        # stable
    embed_kg_nodes=False,   # faster
)
```

```python
import os

BATCH_SAVE_DIR = "/workspace/batch_nodes"
INDEX_PERSIST_DIR = "/workspace/graph_storage"

# Only create if directory does not exist
if not os.path.exists(BATCH_SAVE_DIR):
    os.makedirs(BATCH_SAVE_DIR)

if not os.path.exists(INDEX_PERSIST_DIR):
    os.makedirs(INDEX_PERSIST_DIR)
```

```python
# -----------------------------
# Identify batches already saved
# -----------------------------
existing_batches = set()
for fname in os.listdir(BATCH_SAVE_DIR):
    if fname.startswith("batch_") and fname.endswith(".pkl"):
        batch_index = int(fname.split("_")[1].split(".")[0])
        existing_batches.add(batch_index)

print(f"Found {len(existing_batches)} existing batches. They will be skipped.")
```

```python
# -----------------------------
# Batch insertion loop
# -----------------------------
for i in tqdm(range(0, len(nodes), BATCH_SIZE)):
    if i in existing_batches:
        print(f"Skipping already processed batch {i}")
        continue

    batch_nodes = nodes[i:i + BATCH_SIZE]
    print(f"Inserting batch starting at index {i} ({len(batch_nodes)} nodes)...")

    try:
        index.insert_nodes(batch_nodes)

        # Save the batch to disk
        batch_file = os.path.join(BATCH_SAVE_DIR, f"batch_{i}.pkl")
        with open(batch_file, "wb") as f:
            pickle.dump(batch_nodes, f)

        # Persist the full index
        index.storage_context.persist(persist_dir=INDEX_PERSIST_DIR)
        print(f"Batch {i} completed and index persisted.")

    except Exception as e:
        print(f"Batch {i} failed with error: {e}")
        continue
```
