```python
from llama_index.llms.ollama import Ollama

llm = Ollama(
    model="llama3.3:70b",
    base_url="http://192.168.200.75:7138",
    request_timeout=1800.0,   # 30 minutes
    temperature=0.0,
)
```

```python
response = llm.complete("Hi")
print(response.text)
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)
```

```python
from llama_index.core.node_parser import SentenceSplitter

splitter = SentenceSplitter(
    chunk_size=600,
    chunk_overlap=80,
)

nodes = splitter.get_nodes_from_documents(documents)

print("Total chunks:", len(nodes))
```

```python
from llama_index.core import PropertyGraphIndex

# Build from documents
index = PropertyGraphIndex.from_documents(
    documents,
    llm=llm,
    embed_model=embed_model,
    show_progress=True,
)
```

```python
from llama_index.core import PropertyGraphIndex
from llama_index.core.graph_stores import SimplePropertyGraphStore

graph_store = SimplePropertyGraphStore()

index = PropertyGraphIndex(
    nodes=[],
    property_graph_store=graph_store,
    llm=llm,
    embed_model=embed_model,
    use_async=False,          # stable
    embed_kg_nodes=False,     # faster
)
```

```python
from tqdm import tqdm

BATCH_SIZE = 20  # reduce slightly for safety

for i in tqdm(range(0, len(nodes), BATCH_SIZE)):
    batch_nodes = nodes[i:i+BATCH_SIZE]
    index.insert_nodes(batch_nodes)

print("Graph building completed.")
```
